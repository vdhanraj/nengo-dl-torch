# From PyTorch to Nengo DL

This tutorial is aimed at PyTorch users who want to understand how to build equivalent networks in Nengo and NengoDL. We will walk through building familiar network architectures, then show how to swap in spiking neuron dynamics — bridging the gap between standard deep learning and neuromorphic computing.

NengoDL combines Nengo's neural simulation framework with PyTorch's automatic differentiation and GPU support. If you already know PyTorch, many concepts will feel familiar, but expressed through Nengo's neuron-population-centric API.

```python
%matplotlib inline
import warnings
import matplotlib.pyplot as plt
import nengo
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import nengo_dl
warnings.simplefilter("ignore")
seed = 0
np.random.seed(seed)
torch.manual_seed(seed)
```

## Building equivalent networks in PyTorch and Nengo

In PyTorch, you define a network by subclassing `nn.Module` and stacking layers explicitly in `forward()`. In Nengo, a network is described as populations of neurons (`nengo.Ensemble`) connected by `nengo.Connection` objects.

The key correspondence:

| PyTorch | Nengo |
|---------|-------|
| `nn.Linear` + `F.relu` | `nengo.Ensemble` with `neuron_type=nengo.RectifiedLinear()` |
| `nn.Linear` weight matrix | `nengo.Connection` `transform` parameter |
| `nn.Module.forward()` | Simulation loop via `nengo_dl.Simulator` |

Below we define the same shallow autoencoder in both frameworks.

```python
n_in = 784
n_hidden = 64
minibatch_size = 50

# PyTorch autoencoder
class Autoencoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Linear(n_in, n_hidden)
        self.decoder = nn.Linear(n_hidden, n_in)
    def forward(self, x):
        x = F.relu(self.encoder(x))
        x = F.relu(self.decoder(x))
        return x

pt_model = Autoencoder()
dummy_out = pt_model(torch.ones(minibatch_size, n_in))
print("PyTorch autoencoder output shape:", dummy_out.shape)
```

## Equivalent Nengo Network

The Nengo autoencoder mirrors the PyTorch version:

- `nengo.Ensemble` with `neuron_type=nengo.RectifiedLinear()` corresponds to a linear layer followed by ReLU activation.
- `nengo.Connection` between `.neurons` attributes corresponds to the weight matrix.
- `synapse=None` means no temporal filtering on the connection — the output is passed through instantaneously, matching the behaviour of a standard PyTorch `Linear` layer.

A `nengo.Probe` records the output of a population, analogous to capturing a tensor in PyTorch.

```python
with nengo.Network() as auto_net:
    # Input node
    nengo_a = nengo.Node(np.zeros(n_in))
    
    # Encoder: n_in -> n_hidden with ReLU neurons
    # Connecting to .neurons requires explicit (n_neurons_post, size_pre) transform
    nengo_b = nengo.Ensemble(n_hidden, 1, neuron_type=nengo.RectifiedLinear(), seed=0)
    enc_weights = np.random.randn(n_hidden, n_in).astype(np.float32) * 0.1
    nengo.Connection(nengo_a, nengo_b.neurons, transform=enc_weights, synapse=None)
    
    # Decoder: n_hidden -> n_in with ReLU neurons
    nengo_c = nengo.Ensemble(n_in, 1, neuron_type=nengo.RectifiedLinear(), seed=1)
    dec_weights = np.random.randn(n_in, n_hidden).astype(np.float32) * 0.1
    nengo.Connection(nengo_b.neurons, nengo_c.neurons, transform=dec_weights, synapse=None)
    
    p_c = nengo.Probe(nengo_c.neurons)

with nengo_dl.Simulator(network=auto_net, minibatch_size=minibatch_size) as sim:
    sim.run_steps(1, data={nengo_a: np.ones((minibatch_size, 1, n_in), dtype=np.float32)})
    print("NengoDL autoencoder output shape:", sim.data[p_c].shape)
```

## Rate vs. Spiking Neurons

One of the most powerful features of Nengo is the ability to seamlessly switch between **rate-coded** and **spiking** neuron models.

- **Rate neurons** (`nengo.RectifiedLinear`, `nengo.LIFRate`): output a continuous firing rate at each timestep. This is analogous to standard ReLU activations in PyTorch — the output is a smooth, real-valued signal.
- **Spiking neurons** (`nengo.SpikingRectifiedLinear`, `nengo.LIF`): emit discrete all-or-nothing spikes. Information is encoded in the *timing* and *rate* of spikes. This is the biologically realistic model and is required for neuromorphic hardware.

The key insight: a network trained with rate neurons can often be converted directly to spiking neurons with minimal loss in accuracy, enabling deployment to neuromorphic chips.

```python
with nengo.Network(seed=0) as net:
    a = nengo.Node(lambda t: np.sin(2 * np.pi * t))
    
    # Rate neurons (RectifiedLinear = ReLU, continuous output)
    b_rate = nengo.Ensemble(10, 1, neuron_type=nengo.RectifiedLinear(), seed=2)
    nengo.Connection(a, b_rate)
    
    # Spiking neurons (SpikingRectifiedLinear, discrete spikes)
    b_spike = nengo.Ensemble(10, 1, neuron_type=nengo.SpikingRectifiedLinear(), seed=2)
    nengo.Connection(a, b_spike)
    
    p_a = nengo.Probe(a)
    p_rate = nengo.Probe(b_rate.neurons)
    p_spike = nengo.Probe(b_spike.neurons)

with nengo_dl.Simulator(net) as sim:
    sim.run_steps(1000)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))
axes[0].plot(sim.trange(), sim.data[p_a])
axes[0].set_title("Input signal"); axes[0].set_xlabel("time (s)")

axes[1].plot(sim.trange(), sim.data[p_rate])
axes[1].set_title("Rate neurons (firing rates)"); axes[1].set_xlabel("time (s)")

# Raster-style: show spike events as vertical lines
spike_data = sim.data[p_spike]  # (1000, 10)
t = sim.trange()
for neuron_idx in range(spike_data.shape[1]):
    spike_times = t[spike_data[:, neuron_idx] > 0]
    axes[2].vlines(spike_times, neuron_idx - 0.4, neuron_idx + 0.4, lw=0.8)
axes[2].set_title("Spiking neurons (spike raster)"); axes[2].set_xlabel("time (s)")
axes[2].set_ylabel("Neuron index")
plt.tight_layout()
plt.savefig("rate_vs_spiking.png", dpi=80)
plt.show()
print("Rate neuron max firing rate:", sim.data[p_rate].max())
```

## Filtering spike trains

Raw spike trains are discrete and noisy. To recover a smooth signal approximating the underlying firing rate, we apply a **synaptic low-pass filter** — a leaky integrator with time constant τ.

In biological neurons this corresponds to the decay of post-synaptic potentials. In NengoDL, the `nengo.Lowpass` synapse can be applied either as a probe synapse (online, during simulation) or post-hoc to recorded data using `filt()`.

Longer τ produces smoother estimates but introduces more temporal lag. There is a fundamental trade-off between smoothness and temporal precision.

```python
# Apply a synaptic low-pass filter to smooth the spike trains
tau = 0.05
filt = nengo.Lowpass(tau)
filtered_spikes = filt.filt(sim.data[p_spike], dt=0.001)

plt.figure(figsize=(10, 4))
plt.plot(sim.trange(), filtered_spikes)
plt.title(f"Filtered spike trains (\u03c4={tau}s)")
plt.xlabel("time (s)")
plt.ylabel("Filtered firing rate")
plt.tight_layout()
plt.savefig("filtered_spikes.png", dpi=80)
plt.show()
```

## Embedding PyTorch Modules with TorchNode

NengoDL provides `nengo_dl.TorchNode` to embed any PyTorch `nn.Module` (or callable) directly inside a Nengo network. This gives you full access to PyTorch's ecosystem — custom layers, pretrained models, learned parameters — while still running inside the Nengo simulation loop.

Key properties of `TorchNode`:
- The module receives input as a 2D tensor of shape `(batch_size, n_features)`.
- Gradients flow through it, so parameters are trainable with `sim.fit()`.
- `shape_in` and `shape_out` must match the module's expected input/output feature dimensions.

Below we compare a standard `nengo.Node` (using NumPy) with a `TorchNode` wrapping a PyTorch module that computes the same function.

```python
# Define a custom PyTorch module
class ScaledExp(nn.Module):
    """Computes scale * exp(x) where scale is a learned parameter."""
    def __init__(self, scale=1.0):
        super().__init__()
        self.log_scale = nn.Parameter(torch.tensor(np.log(scale), dtype=torch.float32))
    def forward(self, x):
        return torch.exp(self.log_scale) * torch.exp(x)

with nengo.Network(seed=0) as torch_net:
    inp = nengo.Node([0.0])
    
    # Standard numpy node for comparison
    exp_np = nengo.Node(lambda t, x: np.exp(x), size_in=1)
    nengo.Connection(inp, exp_np, synapse=None)
    p_np = nengo.Probe(exp_np, synapse=0.01)
    
    # TorchNode with PyTorch module
    exp_module = ScaledExp(scale=1.0)
    exp_torch = nengo_dl.TorchNode(exp_module, shape_in=(1,), shape_out=(1,))
    nengo.Connection(inp, exp_torch, synapse=None)
    p_torch = nengo.Probe(exp_torch, synapse=0.01)
    
    p_in = nengo.Probe(inp)

# Run with a sinusoidal input
t_total = 1.0
n_steps = int(t_total / 0.001)
t_arr = np.linspace(0, t_total, n_steps)
inp_data = np.sin(2 * np.pi * t_arr).reshape(1, n_steps, 1).astype(np.float32)

with nengo_dl.Simulator(torch_net, seed=0) as sim:
    sim.run_steps(n_steps, data={inp: inp_data})

plt.figure(figsize=(10, 4))
t = sim.trange()
inp_vals = sim.data[p_in]
plt.plot(t, np.exp(np.sin(2*np.pi*t)), label="True exp(sin(t))", lw=2)
plt.plot(t, sim.data[p_np], label="nengo.Node (numpy)", linestyle="--")
plt.plot(t, sim.data[p_torch], label="TorchNode (PyTorch)", linestyle=":")
plt.title("TorchNode vs nengo.Node")
plt.xlabel("time (s)")
plt.legend()
plt.tight_layout()
plt.savefig("torchnode_comparison.png", dpi=80)
plt.show()
```

## Training with NengoDL

NengoDL exposes a Keras-like training interface that uses PyTorch optimizers under the hood. The workflow mirrors what you'd do in PyTorch:

1. **Compile** the simulator with an optimizer and loss function: `sim.compile(optimizer='adam', loss={probe: 'mse'})`.
2. **Fit** on training data: `sim.fit({input_node: x_data}, {output_probe: y_data}, epochs=N)`.
3. **Evaluate**: `sim.evaluate({input_node: x_test}, {output_probe: y_test})`.

Data shapes follow the convention `(n_samples, n_steps, n_features)` — the extra `n_steps` dimension reflects that Nengo networks process time-series data.

Below we train a two-layer network to approximate `sin(x²)`.

```python
# Build a simple function approximator: sin(x^2)
with nengo.Network(seed=seed) as func_net:
    inpt = nengo.Node(np.zeros(1))
    
    # Two ensembles to compute sin(x^2)
    ens0 = nengo.Ensemble(50, 1, neuron_type=nengo.LIFRate(), seed=0)
    ens1 = nengo.Ensemble(50, 1, neuron_type=nengo.LIFRate(), seed=1)
    outpt = nengo.Node(size_in=1)
    
    nengo.Connection(inpt, ens0, synapse=None)
    nengo.Connection(ens0, ens1, function=np.square, synapse=None)
    nengo.Connection(ens1, outpt, function=np.sin, synapse=None)
    
    outpt_p = nengo.Probe(outpt, synapse=0.01)
    outpt_p_nofilt = nengo.Probe(outpt, synapse=None)

# Show network output BEFORE training
with nengo_dl.Simulator(func_net, seed=seed) as sim:
    t_input = np.linspace(-1, 1, 1000).reshape(1, 1000, 1).astype(np.float32)
    sim.run_steps(1000, data={inpt: t_input})
    
    plt.figure(figsize=(10, 4))
    t = sim.trange()
    plt.plot(t, t_input[0], label="x", alpha=0.7)
    plt.plot(t, np.sin(t_input[0]**2), label="sin(x\u00b2) target", lw=2)
    plt.plot(t, sim.data[outpt_p], label="network output (before training)")
    plt.title("Network output before training")
    plt.xlabel("time (s)")
    plt.legend()
    plt.tight_layout()
    plt.savefig("before_training.png", dpi=80)
    plt.show()
```

```python
# Training data: random x values in [-1, 1]
batch_size = 2048
n_steps_train = 1
vals = np.random.uniform(-1, 1, size=(batch_size, n_steps_train, 1)).astype(np.float32)
targets = np.sin(vals**2).astype(np.float32)

# Train
with nengo_dl.Simulator(func_net, minibatch_size=32, seed=seed) as sim:
    sim.compile(optimizer='adam', loss={outpt_p_nofilt: 'mse'})
    history = sim.fit(
        {inpt: vals},
        {outpt_p_nofilt: targets},
        n_steps=n_steps_train,
        epochs=20,
        verbose=1,
    )

plt.figure(figsize=(8, 4))
plt.plot(history['loss'])
plt.title("Training loss")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.tight_layout()
plt.savefig("training_loss.png", dpi=80)
plt.show()
```

```python
# Show output AFTER training
with nengo_dl.Simulator(func_net, seed=seed) as sim:
    t_input = np.linspace(-1, 1, 1000).reshape(1, 1000, 1).astype(np.float32)
    sim.run_steps(1000, data={inpt: t_input})
    
    plt.figure(figsize=(10, 4))
    t = sim.trange()
    plt.plot(t, t_input[0], label="x", alpha=0.7)
    plt.plot(t, np.sin(t_input[0]**2), label="sin(x\u00b2) target", lw=2)
    plt.plot(t, sim.data[outpt_p], label="network output (after training)")
    plt.title("Network output after training")
    plt.xlabel("time (s)")
    plt.legend()
    plt.tight_layout()
    plt.savefig("after_training.png", dpi=80)
    plt.show()
```

## Cross-Platform Execution

One of Nengo's strongest advantages over plain PyTorch is **backend portability**. A network defined once in Nengo can run on:

- `nengo.Simulator` — standard CPU simulation (pure Python/NumPy, no external dependencies)
- `nengo_dl.Simulator` — GPU-accelerated simulation with full PyTorch gradient support (this library)
- `nengo_loihi.Simulator` — Intel's Loihi neuromorphic chip (ultra-low power spiking hardware)
- `nengo_spinnaker.Simulator` — Manchester's SpiNNaker massively-parallel neuromorphic platform

All backends consume the same `nengo.Network` definition. You only change the simulator class. This makes it straightforward to prototype in software and deploy to neuromorphic hardware without rewriting the model.

Note: `nengo.Simulator` uses a different data convention — it does not have the batch dimension, so input data shapes are `(n_steps, n_features)` instead of `(1, n_steps, n_features)`.

```python
# The same Nengo network can run on multiple backends — just swap the simulator class.
# nengo.Simulator uses the same network but does NOT support the data= / batching API.

with nengo.Simulator(func_net, seed=seed) as sim:
    sim.run_steps(1000)  # uses the Node's built-in zeros output
    print("nengo.Simulator - standard CPU run")
    print(f"  Output shape: {sim.data[outpt_p].shape}")

# with nengo_dl.Simulator(func_net, seed=seed) as sim:
#     ...  # GPU-accelerated PyTorch — nengo_dl.Simulator (this library)

# with nengo_loihi.Simulator(func_net, seed=seed) as sim:
#     ...  # Intel Loihi neuromorphic chip
```
